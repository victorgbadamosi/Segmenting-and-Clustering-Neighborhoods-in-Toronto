```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

Scraping of the dataset from the wikipedia page

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
data = response.text
soup = BeautifulSoup(data, 'html.parser')
wiki_table = soup.find('table')
df = pd.read_html(str(wiki_table))[0]
df.head()
```

## Data preparation

Cleaning the data, whereby ignoring the 'Borough' contains 'Not assigned' cells

```python
df2 = df[df['Borough'].str.contains('Not assigned') == False].reset_index()
df2.head(10)
```

Combining the rows in which postal code and borough

```python
df3 = df2.groupby(['Postal Code', 'Borough']) ['Neighbourhood'].apply(', '.join).reset_index()
df3.head(10)
```
